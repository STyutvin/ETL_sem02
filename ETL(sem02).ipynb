```python
import pandas as pd
```

```python
df = pd.read_csv('/content/fifa_s2.csv')
df.head(10)
```

```python
df.info()
```

```python
# удаление полных дубликатов

df.drop_duplicates()
```

Так как количество строк не изменилось, то полные дубликаты отстутствуют.

```python
# удаление дубликатов в столбцах Name, Club, Joined

df.drop_duplicates(subset=['Name', 'Club'])
```

```python
# добавление столбца с разбиением возраста

def age_range(n):
  age_range = []
  for i, row in n.iterrows():
    if row['Age'] < 20:
      age_range.append('<20')
    elif 20 <= row['Age'] < 30:
      age_range.append('20-30')
    elif 30 <= row['Age'] <= 36:
      age_range.append('30-36')
    elif row['Age'] > 36:
      age_range.append('>36')
  n['Age range'] = age_range
  return n

age_range(df)
```

```python
# подсчет количества футболистов в каждой категории

(df['Age range']).value_counts()
```
